# Glassdoor data scraping with beautifulsoup
The main principle of the codes below is that they reach the given website via url and then scrapes the data of company names, job names, locations of the jobs and salary ranges. After getting data from the website, code transforms the data to the pandas dataframe.

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
r = requests.get("https://www.glassdoor.com/Job/united-states-electrical-engineer-jobs-SRCH_IL.0,13_IN1_KO14,33.htm?includeNoSalaryJobs=false")
soup = bs(r.content)
new_box = soup.body.script.next_sibling.ul

company_names = []
job_names = []
locations = []
salary_ranges = []
# company_name = new_box.find_all("div", {"class":"d-flex justify-content-between align-items-start"})[0].span.string
for i in new_box:
    try:
        company_name = i.find_all("div", {"class":"d-flex justify-content-between align-items-start"})[0].span.string
        company_names.append(company_name)
    except IndexError:
        company_name = 'NULL'
        company_names.append(company_name)
    try:
        job_name = i.find_all("a", {"data-test":"job-link"})[0].span.string
        job_names.append(job_name)
    except IndexError:
        job_name = 'NULL'
        job_names.append(job_name)
    try:
        location = i.find_all("span", {"class":"css-1buaf54 pr-xxsm job-search-key-iii9i8 e1rrn5ka4"})[0].string
        locations.append(location)
    except IndexError:
        location = 'NULL'
        locations.append(location)
    try:
        salary_range = i.find_all("span", {"data-test":"detailSalary"})[0].get_text()
        salary_ranges.append(salary_range)
    except IndexError:
        salary_range = 'NULL'
        salary_ranges.append(salary_range)
        
    
print(company_names)
print(len(company_names))
print(job_names)
print(len(job_names))
print(locations)
print(len(locations))
print(salary_ranges)
print(len(salary_ranges))

df = pd.DataFrame(list(zip(job_names, locations, salary_ranges)), index = company_names, columns = 
                       ['Job Names','Locations','Salary Wages'])
df
```
