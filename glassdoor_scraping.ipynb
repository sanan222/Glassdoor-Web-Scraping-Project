```python
from selenium import webdriver
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

website = "https://www.glassdoor.com/Job/us-electrical-engineer-jobs-SRCH_IL.0,2_IN1_KO3,22.htm"
r = requests.get(website)
soup = bs(r.content)

path = "D:\Downloads\chromedriver\chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get(website)

page_info = soup.find_all("div", {"class":"paginationFooter"})[0].string.split()
current_page = int(page_info[1])
number_of_pages = int(page_info[3])

company_names = []
job_names = []
locations = []
salary_ranges = []

while current_page <= number_of_pages: 
    try:
        pop_up_x_button = driver.find_element(By.XPATH, '//span [@alt = "Close"]')
        pop_up_x_button.click()
        print("Pop up is found in the webpage {}, and closed".format(current_page))
        driver.implicitly_wait(5)
    except:
        print("There are no pop ups on webpage {}".format(current_page))
    finally:
        post_boxes = soup.body.script.next_sibling.ul
        for elements in post_boxes:
            try:
                company_name = elements.find_all("div", {"class":"d-flex justify-content-between align-items-start"})[0].span.string
                company_names.append(company_name)
            except IndexError:
                company_name = 'NULL'
                company_names.append(company_name)
            try:
                job_name = elements.find_all("a", {"data-test":"job-link"})[0].span.string
                job_names.append(job_name)
            except IndexError:
                job_name = 'NULL'
                job_names.append(job_name)
            try:
                location = elements.find_all("span", {"class":"css-1buaf54 pr-xxsm job-search-key-iii9i8 e1rrn5ka4"})[0].string
                locations.append(location)
            except IndexError:
                location = 'NULL'
                locations.append(location)
            try:
                salary_range = elements.find_all("span", {"data-test":"detailSalary"})[0].get_text()
                salary_ranges.append(salary_range)
            except IndexError:
                salary_range = 'NULL'
                salary_ranges.append(salary_range)
        
        next_button = driver.find_element(By.XPATH, '//button [@class = "nextButton css-1hq9k8 e13qs2071"]')
        next_button.click()
        print("Next button clicked")
        driver.implicitly_wait(5)
        
        website = driver.current_url
        print("Webpage url is: {}".format(website))
        soup = bs(r.content)
        
    current_page += 1

print(company_names)
print(len(company_names))
print(job_names)
print(len(job_names))
print(locations)
print(len(locations))
print(salary_ranges)
print(len(salary_ranges))

df = pd.DataFrame(list(zip(job_names, locations, salary_ranges)), index = company_names, columns = 
                       ['Job Names','Locations','Salary Wages'])

'''
df.to_csv('company_info.csv', index=False)
df.to_excel('company_info.xls', index=False)
'''
```
